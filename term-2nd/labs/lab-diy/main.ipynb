```python
import pandas as pd
import numpy as np
```

```python
# Read The File
df = pd.read_csv("Airlines.csv")
df.head()
```

1. In ra tên các cột có chứa phần tử na, điền vào các vị trí na mean của cột đó.

```python
# Print the first 4 rows contain na
df[df.isna().any(axis = 1)].head(4)
```

```python
# Print the name of na cols
nan_cols = [i for i in df.columns if df[i].isna().any()]
nan_cols
```

```python
# Use this if you want to fill the entire dataframe
#df = df.fillna(0)
# Or this when filling specific column
df["Arrival Delay in Minutes"] = df["Arrival Delay in Minutes"].fillna(df["Arrival Delay in Minutes"].mean())
```

```python
# Check for the amount of nan cols
df.isna().any().sum()
```

2. In ra 3 chuyến bay cuối với "Flight Distance" < 800

```python
# Print the last 3 flights with less than 800 in "Flight Distance"
df[df["Flight Distance"] < 800].tail(3)
```

3. In ra các chuyến bay thỏa: "Inflight wifi service" > 4; "Departure Delay in Minutes" >= 20 và "Type of Travel" == "Personal Travel"

```python
'''
Print all the flights that satisfy 
("Inflight wifi service" > 4, 
"Departure Delay in Minutes" >= 20, 
"Type of Travel" == "Personal Travel")
'''
df[ (df["Inflight wifi service"] > 4) & (df["Departure Delay in Minutes"] >= 20) & (df["Type of Travel"] == "Personal Travel") ]
```

4. Thêm cột năm sinh NGAY SAU cột tuổi

```python
# Add a col with birthyear of passengers right after age
import datetime as time
curYear = time.datetime.now().year
df.insert(df.columns.get_loc("Age") + 1, 'Birth Year', curYear - df["Age"])
df.head()
```

5. In ra các chuyến bay bị delay từ 30' trở lên

```python
# Print all the passenger with the total amount of delay >= 30
df[ (df["Departure Delay in Minutes"] + df["Arrival Delay in Minutes"]) >= 30].head()
```

6. In ra Flight Distance Average/ Max/ Min theo từng class trong cột "Class"

```python
# Print the average/ max/ min flight distance (in int) by each categories of "Class"
classes = df["Class"].unique()
for c in classes:
    print( c + " - Average: ", (df[df["Class"] == c])["Flight Distance"].mean().astype(int))
    print( c + " - Max: ", (df[df["Class"] == c])["Flight Distance"].max().astype(int))
    print( c + " - Min: ", (df[df["Class"] == c])["Flight Distance"].min().astype(int))
```

7. In tần số được cho điểm của cột "Seat comfort", Sau đó convert ra dạng %

```python
# Print the frequency of "Seat comfort" grading
print("The frequencies of seat comfort grading are: ")
print(df["Seat comfort"].value_counts())
percentages = df.groupby("Seat comfort").size() / len(df) * 100
print(percentages)
```

8. In ra Flight Distance Average và độ lệch chuẩn của khách loyal từ 20 đến 40 tuổi

```python
# Print the average flight distance of Loyal Customers age between 20 and 40
df_target_ages = df[ (df["Age"] >= 20) & (df["Age"] <= 40) & (df["Customer Type"] == "Loyal Customer")]
print(df_target_ages["Flight Distance"].mean().astype(float))
print(df_target_ages["Flight Distance"].std().astype(float))
```

9. In ra % satisfaction của nam/ nữ (tổng 100 %), làm tròn 2 chữ số tp

```python
# Print percentages of satisfaction anmong genders
percentages = df.groupby(["satisfaction", "Gender"]).size() / len(df) * 100
print(round(percentages,2))
```

10. In ra Flight Distance Average, group theo 2 tiêu chí "Type of Travel" và "Online boarding"

```python
# Print the average flight distance group by "Type of Travel" & "Online boarding" (Please use groupby)
df.groupby(["Type of Travel", "Online boarding"])["Flight Distance"].mean().astype(int)
```

11. Xem các cột từ "Inflight wifi service" đến "Cleanliness" là các cột điểm đến 5
    Tính tổng điểm cho mỗi chuyến bay (Ko được phép hardcode bằng cách ghi df["1"] + df["2"] + ...)

```python
# Consider cols starting from "Inflight wifi service" to "Cleanliness" as a grading system
# Add a score col (sum of all scores) at the front of "Inflight wifi service""(no hard coding allowed)
idx = df.columns.get_loc("Inflight wifi service")
df.insert(idx,"Scoring", df.loc[:, "Inflight wifi service":"Cleanliness"].sum(axis=1))
df.head(5)
```

12. Format lại "Customer Type" theo Capitalize Each Word, dùng Apply.

```python
# Format the content inside "Customer Type" with the <Capitalize Each Word> Rule using apply
df["Customer Type"] = df["Customer Type"].apply(lambda x: x.capitalize())
df.head()
```

13. Phân loại flights distance của mỗi chuyến ra thành một cột Flight Distance Categories
Theo luật sau:
< 1000: Short
1000 <= x <= 2500: Medium
2500 >: Long

```python
# Categories flights distance into a new col name "Flight Distance Categories" right after Flight Distance
'''
< 1000: Short
1000 <= x <= 2500: Medium
2500 >: Long
'''
df["Flight Distance Categories"] = np.nan
df.loc[df["Flight Distance"] < 1000, ["Flight Distance Categories"]] = 'Short'
df.loc[df["Flight Distance"] > 2500, ["Flight Distance Categories"]] = 'Long'
df["Flight Distance Categories"].fillna("Medium", inplace=True)
df.head()
```

14. Merge 2 file Airlines-Fee.xlsx và Airlines.csv sao cho rate tương ứng với class
Thêm cột receipt = rate * flight distance.

```python
# Calculating the receipt for each passengers by multiply the number in Airlines-Fee with the Flight Distance:
df_price = pd.read_excel("Airlines-Fee.xlsx")
df_price
```

```python
df_merge = df.merge(df_price, left_on="Class", right_on="Class")
df_merge["Receipt"] = df_merge["Flight Distance"] * df_merge["Rate"]
df_merge.head(5)
```

15. Format lại tất cả các TÊN CỘT theo kiểu Capitalized toàn bộ (df_merge)

```python
df_merge.rename(columns=lambda x: x.upper(), inplace=True)
df_merge.head()
```

16. Sort dataframe đã merge theo SCORING -> RECEIPT, giảm dần

```python
df_merge.sort_values(by = ["SCORING", "RECEIPT"], ascending=False, inplace=True)
df_merge.head(10)
```
